```python
import pandas as pd

# Charger le CSV original
df = pd.read_csv("out-Archive.csv")

# Convertir le temps en format ISO 8601 (ex: 2019-06-08T03:55:45Z)
df['time'] = pd.to_datetime(df['Time'], format='%H:%M:%S %d-%b-%Y').dt.strftime('%Y-%m-%dT%H:%M:%SZ')

# Renommer profondeur et température
df = df.rename(columns={
    'Temperature': 'temperature',
    'Depth': 'pressure',
    'Light Level':'light',
})

# Réorganiser les colonnes
colonnes = [
    'time', 'temperature', 'pressure',
    'light', 'Ax', 'Ay', 'Az', 'Dry', 'Aux',
    'One Minute Light Level', 'Smoothed Light Level', 'Events'
]
df = df[colonnes]

# 🔥 Convertir toutes les colonnes numériques en float
numeric_cols = [
    'temperature', 'pressure', 'light', 'Ax', 'Ay', 'Az',
    'Dry', 'Aux', 'One Minute Light Level', 'Smoothed Light Level'
]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Supprimer les lignes entièrement vides (facultatif mais utile)
df = df.dropna(subset=['temperature', 'pressure'])

# Sauvegarder le CSV final
df.to_csv("dst_.csv", index=False)

print(df.dtypes)  # Vérifier les types si tu veux
```
